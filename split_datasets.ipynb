```python
import os

# 데이터셋 분리

base_dir = './datasets'

origin_dir = os.path.join(base_dir, 'origin')
train_dir = os.path.join(base_dir, 'train')
test_dir = os.path.join(base_dir, 'test')

train_0_front_dir = os.path.join(train_dir, '0_front')
train_1_back_dir = os.path.join(train_dir, '1_back')
train_1_front_dir = os.path.join(train_dir, '1_front')
train_2_back_dir = os.path.join(train_dir, '2_back')
train_2_front_dir = os.path.join(train_dir, '2_front')
train_5_front_dir = os.path.join(train_dir, '5_front')
train_ILU_dir = os.path.join(train_dir, 'ILU')

test_0_front_dir = os.path.join(test_dir, '0_front')
test_1_back_dir = os.path.join(test_dir, '1_back')
test_1_front_dir = os.path.join(test_dir, '1_front')
test_2_back_dir = os.path.join(test_dir, '2_back')
test_2_front_dir = os.path.join(test_dir, '2_front')
test_5_front_dir = os.path.join(test_dir, '5_front')
test_ILU_dir = os.path.join(test_dir, 'ILU')

save_dir = os.path.join(base_dir, 'saved_models')
```

```python
# train dir 생성
if not os.path.isdir(train_0_front_dir) :
    os.mkdir(train_0_front_dir)
    
if not os.path.isdir(train_1_back_dir) :
    os.mkdir(train_1_back_dir)
    
if not os.path.isdir(train_1_front_dir) :
    os.mkdir(train_1_front_dir)
    
if not os.path.isdir(train_2_back_dir) :
    os.mkdir(train_2_back_dir)
    
if not os.path.isdir(train_2_front_dir) :
    os.mkdir(train_2_front_dir)
    
if not os.path.isdir(train_5_front_dir) :
    os.mkdir(train_5_front_dir)
    
if not os.path.isdir(train_ILU_dir) :
    os.mkdir(train_ILU_dir)

    
# test dir 생성
if not os.path.isdir(test_0_front_dir) :
    os.mkdir(test_0_front_dir)

if not os.path.isdir(test_1_back_dir) :
    os.mkdir(test_1_back_dir)
    
if not os.path.isdir(test_1_front_dir) :
    os.mkdir(test_1_front_dir)
    
if not os.path.isdir(test_2_back_dir) :
    os.mkdir(test_2_back_dir)
    
if not os.path.isdir(test_2_front_dir) :
    os.mkdir(test_2_front_dir)
    
if not os.path.isdir(test_5_front_dir) :
    os.mkdir(test_5_front_dir)
    
if not os.path.isdir(test_ILU_dir) :
    os.mkdir(test_ILU_dir)


# save dir
if not os.path.isdir(save_dir) :
    os.mkdir(save_dir)
```

```python
import shutil
import random

# test image : 100장
# train image : 나머지 전부
for class_name in sorted(os.listdir(origin_dir)):
    class_imgs = sorted(os.listdir(os.path.join(origin_dir, class_name)))
    random.shuffle(class_imgs)
    
    src_test_imgs = [os.path.join(origin_dir, class_name, img) for img in class_imgs[:100]]
    src_train_imgs = [os.path.join(origin_dir, class_name, img) for img in class_imgs[100:]]
    
    dst_test_imgs = [os.path.join(test_dir, class_name, img) for img in class_imgs[:100]]
    dst_train_imgs = [os.path.join(train_dir, class_name, img) for img in class_imgs[100:]]
    
    for src, dst in zip(src_test_imgs, dst_test_imgs):
        shutil.copy(src, dst)
        
    for src, dst in zip(src_train_imgs, dst_train_imgs):
        shutil.copy(src, dst)
```
